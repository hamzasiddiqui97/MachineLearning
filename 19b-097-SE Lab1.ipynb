```python
#Convert data into integer

from csv import reader

def load_csv(filename):
    dataset = list ()
    with open ( filename ,'r') as file:
        csv_reader = reader(file)
        for row in csv_reader:
            if not row :
                continue
            dataset.append(row)
    return dataset

def str_column_to_int (dataset , column):
    class_values = [row[column] for row in dataset]
    unique = set(class_values)
    lookup = dict()
    for i, value in enumerate (unique):
        lookup[value] = i
    for row in dataset :
        row [column] = lookup[row[column]]
    return lookup

filename = 'diabetes_notepad.txt'
dataset = load_csv( filename )
print(f'Loaded data file name = {filename}\nRows = {len(dataset)}\nColumns = {len(dataset[0])}')

dataset = dataset[1::]
for i in range(len(dataset[0])):
    str_column_to_int(dataset, i)

for i in range(len(dataset)):
    print(dataset[i])
```

```python
#Task1
from csv import reader

def load_csv ( filename ):
    file = open ( filename , "r")
    lines = reader (file)
    dataset = list (lines)
    return dataset

def str_column_to_float(dataset, column):
    for row in dataset:
        row[column] = float(row[column].strip())
        
filename = 'diabetes.csv'
dataset = load_csv(filename)

# for excluding row 1 of the data
dataset = dataset[1::]

for i in range(len(dataset[0])):
    str_column_to_float(dataset, i)

for i in range(len(dataset)):
    print(dataset[i])
```

```python
# task2
def missing_value(dataset, column):
    count = 0
    for i in range(len(dataset[0])):
        str_column_to_int(dataset, i)
  
    for i in dataset:
        if not i[column]:
            count += 1
    return count
    
filename = 'diabetes.csv'
dataset = load_csv(filename)
missing_value(dataset,1)
```

```python
#task 4

def delimiters(filename):
    dataset = []
    file = open ( filename , "r")
    lines = reader ( file , delimiter =' ')
    for line in lines:
        if not line:
            continue
        else:
            dataset.append(line)
    return dataset

filename = 'diabetes.csv'
delimiters(filename)
```
